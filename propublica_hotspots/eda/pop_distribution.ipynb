# Creating plots to show the distribution of the population across the data set
Across grids, hotspots, etc.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
gridspaces_df = pd.read_csv("../data/al_grid.csv")
gridspaces_df
```

```python
hotspots_df = pd.read_csv("../data/al_hotspots.csv")
hotspots_df
```

## Population distribution across all gridsquares

#### Grouping grid spaces within 5 people into each bin

```python
plt.clf()
gridspaces_df_filtered = pd.DataFrame(gridspaces_df)
gridspaces_df_filtered["Total"] = gridspaces_df_filtered["Total"].apply(np.floor).astype(int)
# gridspaces_df_filtered = gridspaces_df_filtered.loc[gridspaces_df_filtered["Total"] < 100]

x, y, _ = plt.hist(gridspaces_df_filtered["Total"], bins=[i - 0.5 for i in range(0, int(gridspaces_df_filtered["Total"].max()) + 1, 5)])
for i in range(20):
    print(f"Count of gridsquares with populations in range [{y[i]}-{y[i + 1]}): {x[i]}")
```

As can be seen, there is a heavy concentration In gridsquares with very small populations. Filtering some of those gridsquares out of the histogram allows us to get a better perspective on the distribution of the remaining values:

Filtering out all populations less than 5

```python
gridspaces_df_filtered = pd.DataFrame(gridspaces_df)
gridspaces_df_filtered["Total"] = gridspaces_df_filtered["Total"].apply(np.floor).astype(int)
gridspaces_df_filtered = gridspaces_df_filtered.loc[gridspaces_df_filtered["Total"] >= 5]

x, y, _ = plt.hist(gridspaces_df_filtered["Total"], bins=[i - 0.5 for i in range(0, int(gridspaces_df_filtered["Total"].max()) + 1, 5)])
```

Populations less than 20

```python
gridspaces_df_filtered = pd.DataFrame(gridspaces_df)
gridspaces_df_filtered["Total"] = gridspaces_df_filtered["Total"].apply(np.floor).astype(int)
gridspaces_df_filtered = gridspaces_df_filtered.loc[gridspaces_df_filtered["Total"] >= 20]

x, y, _ = plt.hist(gridspaces_df_filtered["Total"], bins=[i - 0.5 for i in range(0, int(gridspaces_df_filtered["Total"].max()) + 1, 5)])
```

Populations less than 50

```python
gridspaces_df_filtered = pd.DataFrame(gridspaces_df)
gridspaces_df_filtered["Total"] = gridspaces_df_filtered["Total"].apply(np.floor).astype(int)
gridspaces_df_filtered = gridspaces_df_filtered.loc[gridspaces_df_filtered["Total"] >= 50]

x, y, _ = plt.hist(gridspaces_df_filtered["Total"], bins=[i - 0.5 for i in range(0, int(gridspaces_df_filtered["Total"].max()) + 1, 5)])
```

An alternative way of analyzing the distribution is to see the percentage of grids in Alabama that have populations above a given value:

```python
cutoffs = [5, 20, 50, 100]

for val in cutoffs:
    percentage = round(len(gridspaces_df.loc[gridspaces_df["Total"] >= val]) / len(gridspaces_df) * 100, 2)
    print(f"% of AL grids with populations greater than {val}: {percentage}%")
```

## Population distribution across all hotspots
Binned in groups of 100

```python
hotspots_df["Total"] = hotspots_df["Total"].apply(np.floor).astype(int)

x, y, _ = plt.hist(hotspots_df["Total"], bins=[i - 0.5 for i in range(0, int(hotspots_df["Total"].max()) + 100, 100)])
```

From this plot, we can see we have a major outlier in terms of population.

```python
hotspots_df_filtered = pd.DataFrame(hotspots_df)
hotspots_df_filtered["Total"] = hotspots_df_filtered["Total"].apply(np.floor).astype(int)
hotspots_df_filtered = hotspots_df_filtered[(hotspots_df_filtered["Total"] < 70000) & (hotspots_df_filtered["Total"] > 0)]

x, y, _ = plt.hist(hotspots_df_filtered["Total"], bins=[i - 0.5 for i in range(0, int(hotspots_df_filtered["Total"].max()) + 100, 100)])
```
