# Creating plots to explore racial distributions across grid spaces and hotspots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
gridspaces_df = pd.read_csv("../data/al_grid.csv")
gridspaces_df
```

```python
hotspots_df = pd.read_csv("../data/al_hotspots.csv")
hotspots_df["Total"] = hotspots_df["Total"].apply(np.floor).astype(int)
hotspots_df
```

## Bar charts showing racial demographics across hotspots

```python
def show_race_stack_bar_chart(data, data_name=None, byrow=False):
    race_cols = [
        'White alone',
        'Black or African American alone',
        'American Indian and Alaska Native alone', 
        'Asian alone',
        'Native Hawaiian and Other Pacific Islander alone',
        'Some Other Race alone',
        'Two or More Races', 
        'Hispanic or Latino'
    ]

    percs = []

    if byrow:
        plt.figure(figsize=(10,10))

        al_total = gridspaces_df["Total"].sum()

        names = (data["place"] + " (" + data["Total"].astype(str) + ")").tolist() + [f"All of Alabama ({round(al_total)})"]

        # Assertion: all rows will be converted to bars and placed on same plot
        for col in race_cols:
            all_al = gridspaces_df[col].sum() / al_total

            perc = pd.Series(data[col + " %"].tolist() + [all_al])

            plt.barh(names, perc, 0.5, left=sum(percs))
            percs.append(perc)

        plt.legend([f"{race_cols[i]}" for i in range(len(race_cols))], bbox_to_anchor=(1.1, 1.05))
        plt.show()
    else:
        # Check that data is a DataFrame, simply get total if not (assuming it is a series)
        total = data["Total"].sum() if isinstance(data, pd.DataFrame) else data["Total"]

        plt.figure(figsize=(10,1.5))

        for col in race_cols:
            race_count = data[col].sum() if isinstance(data, pd.DataFrame) else data[col]
            perc = race_count / total

            plt.barh(["", "", ""], perc, 0.1, left=sum(percs))
            percs.append(perc)

        plt.legend([f"{race_cols[i]} ({round(percs[i] * 100, 3)}%)" for i in range(len(race_cols))], bbox_to_anchor=(1.1, 1.05))
        plt.title(data_name)
        plt.ylim(-0.2, 0.2)
        plt.show()
```
