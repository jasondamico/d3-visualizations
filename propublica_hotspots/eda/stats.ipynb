## Percentages of whole of Alabama

#### Percent of AL residents who live in high cancer risk zones

```python
sum(df.loc[df["ilcr"] != 0]["Total"]) / sum(df["Total"])
```

#### Percent of Alabama residents who are white divided by the percentage of residents in high cancer risk areas that are white

```python
white_perc_in_al = sum(df["White alone"]) / sum(df["Total"])
white_perc_in_sacrifice = sum(df.loc[df["ilcr"] != 0]["White alone"]) / sum(df.loc[df["ilcr"] != 0]["Total"])

white_perc_in_al / white_perc_in_sacrifice
```

#### Percent of white residents that live in sacrifice zones

```python
sum(df.loc[df["ilcr"] != 0]["White alone"]) / sum(df["White alone"])
```

```python
def print_race_in_state_and_zones(race):
    sac_zone_percent = round(sum(df.loc[df["ilcr"] != 0]["Total"]) / sum(df["Total"]) * 100, 3)
    sac_zone_percent_race = round(sum(df.loc[df["ilcr"] != 0][race]) / sum(df[race]) * 100, 3)

    print(f"Percent of Alabamans in sacrifice zones: {sac_zone_percent}%")
    print(f"Percent of {race.replace(' alone', '')} Alabamans in sacrifice zones: {sac_zone_percent_race}%")
```

```python
race_cols = df.columns[(df.columns.str.endswith(" %")) & (~df.columns.str.contains(":"))].str.replace(" %", "").tolist()

for col in race_cols:
    print_race_in_state_and_zones(col)
    print()
```
