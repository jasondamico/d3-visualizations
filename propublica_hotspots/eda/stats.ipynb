#### Percent of Alabama residents who are white divided by the percentage of residents in high cancer risk areas that are white

```python
white_perc_in_al = sum(df["White alone"]) / sum(df["Total"])
white_perc_in_sacrifice = sum(df.loc[df["ilcr"] != 0]["White alone"]) / sum(df.loc[df["ilcr"] != 0]["Total"])

white_perc_in_al / white_perc_in_sacrifice
```

#### Percent of white residents that live in sacrifice zones

```python
sum(df.loc[df["ilcr"] != 0]["White alone"]) / sum(df["White alone"])
```

```python
def print_race_in_state_and_zones(race):
    sac_zone_percent = round(sum(df.loc[df["ilcr"] != 0]["Total"]) / sum(df["Total"]) * 100, 3)
    sac_zone_percent_race = round(sum(df.loc[df["ilcr"] != 0][race]) / sum(df[race]) * 100, 3)

    print(f"Percent of Alabamans in sacrifice zones: {sac_zone_percent}%")
    print(f"Percent of {race.replace(' alone', '')} Alabamans in sacrifice zones: {sac_zone_percent_race}%")
```

```python
race_cols = df.columns[(df.columns.str.endswith(" %")) & (~df.columns.str.contains(":"))].str.replace(" %", "").tolist()

for col in race_cols:
    print_race_in_state_and_zones(col)
    print()
```

## Grouping grid squares with cancer risk greater than or equal to the given ILCR

```python
calcs = pd.DataFrame(columns=["# of grids", "ilcr"] + [col for col in filtered_df.columns if "%" in col])

cancer_risks = [None, 0.1, (1 / 75000) * 10000, (1 / 50000) * 10000, (1 / 25000) * 10000, 1]

for risk in cancer_risks:
    if risk == None:
        # Handling for grid squares without cancer risks calculated in ProPublica data
        percentages_dict = filtered_df.loc[:, [col for col in filtered_df.columns if "%" in col]].mean().to_dict()
        percentages_dict["ilcr"] = float("nan")
        percentages_dict["# of grids"] = len(filtered_df)
    else:
        percentages_dict = filtered_df.loc[filtered_df["ilcr"] >= risk, [col for col in filtered_df.columns if "%" in col]].mean().to_dict()
        percentages_dict["ilcr"] = risk
        percentages_dict["# of grids"] = len(filtered_df[filtered_df["ilcr"] >= risk])

    calcs = calcs.append(percentages_dict, ignore_index=True)

calcs
```

## Grouping grid squares with cancer risk in a given range

```python
non_overlapping = pd.DataFrame(columns=["# of grids", "ilcr range"] + [col for col in filtered_df.columns if "%" in col])

cancer_risks = [0, 0.1, (1 / 75000) * 10000, (1 / 50000) * 10000, (1 / 25000) * 10000, 1, 99]

for i in range(len(cancer_risks)):
    risk = cancer_risks[i]

    if risk == 0:
        # Handling for grid squares without cancer risks calculated in ProPublica data
        percentages_dict = filtered_df.loc[:, [col for col in filtered_df.columns if "%" in col]].mean().to_dict()
        percentages_dict["ilcr range"] = float("nan")
        percentages_dict["# of grids"] = len(filtered_df)
    else:
        percentages_dict = filtered_df.loc[(filtered_df["ilcr"] <= risk) & (filtered_df["ilcr"] >= cancer_risks[i - 1]), [col for col in filtered_df.columns if "%" in col]].mean().to_dict()
        percentages_dict["ilcr range"] = f"{cancer_risks[i - 1]}-{risk}"
        percentages_dict["# of grids"] = len(filtered_df[(filtered_df["ilcr"] <= risk) & (filtered_df["ilcr"] >= cancer_risks[i - 1])])

    non_overlapping = non_overlapping.append(percentages_dict, ignore_index=True)

non_overlapping
```
